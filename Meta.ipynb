```python
import pandas as pd

df = pd.read_csv(
    filepath_or_buffer='meta_data.csv', 
    header=None, 
    sep=',')

# df.columns=['GH1', 'GH3', 'GH5','GH9','GH44','GH48','GH10','GH11' ,'GH39','GH43', 'GH67', 'GH16', 'GH31', 'GH2', 'GH26', 'GH36', 'GH88']

df
```

```python
X = df.ix[:,0:21].values
# y = df.ix[:,16].values
print(X)
```

```python
import numpy as np
x = np.matrix(df.ix[:,0:21].values.reshape((17, 21)))
# x.mean(1)
x.std(1, ddof=1)
```

```python
from sklearn.preprocessing import StandardScaler
X_std = StandardScaler(copy=True, with_mean=True, with_std=False).fit_transform(X.T)
X_std = X_std.T
```

```python
print(X_std)
```

```python
X_std2 = X_std/x.std(1, ddof=1)
X_std2.round(2)
```

```python
(X_std2.shape[1]-1)
```

```python
import numpy as np
mean_vec = np.mean(X_std2, axis=1)
# print(mean_vec)
cov_mat = (X_std2).dot((X_std2).T) / (X_std2.shape[1]-1)
print('Covariance matrix \n%s' %cov_mat.round(3))
```

```python
eig_vals, eig_vecs = np.linalg.eig(cov_mat)

print('Eigenvectors \n%s' %eig_vecs.round(2))
print('\nEigenvalues \n%s' %eig_vals.round(2))
```

```python
for ev in eig_vecs:
    np.testing.assert_array_almost_equal(1.0, np.linalg.norm(ev))
print('Everything ok!')
```

```python
# Make a list of (eigenvalue, eigenvector) tuples
eig_pairs = [(np.abs(eig_vals[i]), eig_vecs[:,i]) for i in range(len(eig_vals))]

# Sort the (eigenvalue, eigenvector) tuples from high to low
eig_pairs.sort()
eig_pairs.reverse()

# Visually confirm that the list is correctly sorted by decreasing eigenvalues
print('Eigenvalues in descending order:')
for i in eig_pairs:
    print(i[0].round(2))
    print(i[1].round(2))
```

```python
import plotly.plotly as py
from plotly.graph_objs import *
import plotly.tools as tls
```

```python
tot = sum(eig_vals)
var_exp = [(i / tot)*100 for i in sorted(eig_vals, reverse=True)]
cum_var_exp = np.cumsum(var_exp)

trace1 = Bar(
        x=['PC %s' %i for i in range(1,18)],
        y=var_exp,
        showlegend=False)

trace2 = Scatter(
        x=['PC %s' %i for i in range(1,18)], 
        y=cum_var_exp,
        showlegend=False)

data = Data([trace1, trace2])

layout=Layout(
        yaxis=YAxis(title='Explained Variance(%)'),
        title='Explained Variance by Principal Components',
#         margin=dict(t=50, b=20, r=0, l=60, autoexpand=False),
        font=dict(size=16)
        )

fig = Figure(data=data, layout=layout)
py.iplot(fig)
```

```python
matrix_w = np.hstack((eig_pairs[0][1].reshape(17,1), 
                      eig_pairs[1][1].reshape(17,1)))

print('Matrix W:\n', matrix_w.round(2))
```

```python
matrix_w2 = np.hstack((eig_pairs[0][1].reshape(17,1), 
                      eig_pairs[1][1].reshape(17,1),
                      eig_pairs[2][1].reshape(17,1)))

print('Matrix W2:\n', matrix_w2.round(2))
```

```python
Y = X_std2.T.dot(matrix_w)
Y.round(2)
```

```python
Y2 = X_std2.T.dot(matrix_w2)
Y2.round(2)
```
